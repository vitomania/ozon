```python
import numpy as np
import pandas as pd
import seaborn as sns
from tqdm import tqdm
sns.set()


class GaussianBandit:
    #принимает на вход список из K>=2 чисел std
    def __init__(self, bandit_probs):
        self.bandit_probs = np.array(bandit_probs) # std

    #Функция возвращает число ручек
    def K(self):
        return len(self.bandit_probs)

    #Принимает параметр 0 <= a <= K-1 и возвращает реализацию случайной величины 
    #X c P[X=1] равной среднему значению выигрыша ручки a+1
    def pull(self, a):
        if isinstance(a, (int, np.int64)):
            return np.random.normal(0, self.bandit_probs[a])
        else:
            return np.random.normal([0] * len(a), self.bandit_probs[a])

    #Возвращает текущее значение регрета
    def regret(self, a):
        return np.max(self.bandit_probs) - self.bandit_probs[a]
```

```python
import matplotlib.pyplot as plt

def plot_results(mean_regret1, mean_regret2, params1, params2, regret1, regret2, time_range, ax):
    ax[0, 0].plot(time_range, mean_regret1, color="purple")
    ax[0, 0].scatter(time_range, mean_regret1, color="purple", s=2**5)
    ax[0, 1].hist(regret1, bins="stone", color="purple", edgecolor='white', linewidth=1)
    ax[0, 1].set_title("Histogram of regret (T = 1000)", fontsize=20)
    ax[0, 1].set_xlabel("Regret", fontsize=20)
    ax[1, 0].plot(time_range, mean_regret2, color="purple")
    ax[1, 0].scatter(time_range, mean_regret2, color="purple", s=2**5)
    ax[1, 1].hist(regret2, bins="stone", color="purple", edgecolor='white', linewidth=1)
    ax[1, 1].set_title("Histogram of regret (T = 1000)", fontsize=20)
    ax[1, 1].set_xlabel("Regret", fontsize=20)

    ax[0, 0].set_title("Params = {}".format(np.round(params1, 3)), fontsize=20)
    ax[1, 0].set_title("Params = {}".format(np.round(params2, 3)), fontsize=20)
    ax[0, 0].set_xlabel("T", fontsize=20)
    ax[0, 0].set_ylabel("Expected regret", fontsize=20)
    ax[1, 0].set_xlabel("T", fontsize=20)
    ax[1, 0].set_ylabel("Expected regret", fontsize=20)
    ax[0, 0].grid(which='major', axis='both', linestyle='-', color='white', linewidth=2)
    ax[1, 0].grid(which='major', axis='both', linestyle='-', color='white', linewidth=2)
```

## UCB

```python
def UCB(bandit, T, delta):
    # delta - error probability
    if T < bandit.K():
        raise Exception("T must be greater than number of arms K.")
        
    n = bandit.K()
    mean_rewards = np.zeros(n)
    n_terms = np.zeros(n)
    regret = 0
    arms = np.arange(n)
    for t in range(n, T):
        eq = mean_rewards + np.sqrt(2 * np.log(1 / delta))
        mask = n_terms >= 1
        eq[~mask] = np.inf
        eq[mask] *= np.sqrt(1 / n_terms[mask])
        mask = eq == np.max(eq)
        arm = np.random.choice(arms[mask])
        
        mean_rewards[arm] = (mean_rewards[arm] * n_terms[arm] + bandit.pull(arm)) / (n_terms[arm] + 1)
        n_terms[arm] += 1
        regret += bandit.regret(arm)
        
    return regret

def simulate_UCB(params, time_range, delta):
    mean_regret = []
    bandit = GaussianBandit(params)
    pbar = tqdm(time_range)
    regret = dict()
    for t in pbar:
        pbar.set_description("Processing T = %s" % t)
        regret[t] = []
        for i in range(1, 1000 + 1):
            regret[t].append(UCB(bandit, T=t, delta=delta))

        mean_regret.append(np.mean(regret[t]))
        
    return np.array(mean_regret), regret
```

```python
time_range = range(100, 600, 100)
params1 = [1, 3]
params2 = [0.1, 0.5, 1, 3, 5, 7, 9, 12, 15, 20]
mean_regret1, regret1 = simulate_UCB(params1, time_range, delta=0.5)
mean_regret2, regret2 = simulate_UCB(params2, time_range, delta=0.5)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(25, 15))
plot_results(mean_regret1, mean_regret2, params1, params2, regret1[500], regret2[500], time_range, ax)
```

## Phased UCB (1)

```python
def UCB_first(bandit, T, delta):
    # delta - error probability
    if T < bandit.K():
        raise Exception("T must be greater than number of arms K.")
        
    n = bandit.K()
    mean_rewards = bandit.pull(np.arange(n))
    n_terms = np.ones(n)
    regret = np.sum(bandit.regret(np.arange(n)))
    arms = np.arange(n)
    l = 0
    for t in range(n, T):
        l += 1
        eq = mean_rewards + np.sqrt(2 / n_terms * np.log(1 / delta))
        mask = eq == np.max(eq)
        arm = np.random.choice(arms[mask])
        
        if 2**l + t >= T:
            arr_arms = np.array([arm] * (T - t))
        else:
            arr_arms = np.array([arm] * (2**l))

        mean_rewards[arm] = (
            (mean_rewards[arm] * n_terms[arm] + sum(bandit.pull(arr_arms))) / 
            (n_terms[arm] + len(arr_arms))
        )
        n_terms[arm] += len(arr_arms)   
        regret += sum(bandit.regret(arr_arms))
        t += len(arr_arms)
        if t >= T:
            return regret
        
    return regret

def simulate_UCB_first(params, time_range, delta):
    mean_regret = []
    bandit = GaussianBandit(params)
    pbar = tqdm(time_range)
    regret = dict()
    for t in pbar:
        pbar.set_description("Processing T = %s" % t)
        regret[t] = []
        for i in range(1, 1000 + 1):
            regret[t].append(UCB_first(bandit, T=t, delta=delta))

        mean_regret.append(np.mean(regret[t]))
        
    return np.array(mean_regret), regret
```

## Phased UCB (2)

```python
def UCB_second(bandit, T, alpha, delta):
    # delta - error probability
    # alpha - threshold
    if T < bandit.K():
        raise Exception("T must be greater than number of arms K.")
        
    n = bandit.K()
    mean_rewards = bandit.pull(np.arange(n))
    n_terms = np.ones(n)
    regret = np.sum(bandit.regret(np.arange(n)))
    arms = np.arange(n)
    l = 0
    for t in range(n, T):
        l += 1
        eq = mean_rewards + np.sqrt(2 / n_terms * np.log(1 / delta))
        mask = eq == np.max(eq)
        arm = np.random.choice(arms[mask])
        
        if int(alpha * n_terms[arm]) + t >= T:
            arr_arms = np.array([arm] * (T - t))
        else:
            arr_arms = np.array([arm] * (int(alpha * n_terms[arm])))

        mean_rewards[arm] = (
            (mean_rewards[arm] * n_terms[arm] + sum(bandit.pull(arr_arms))) / 
            (n_terms[arm] + len(arr_arms))
        )
        n_terms[arm] += len(arr_arms)   
        regret += sum(bandit.regret(arr_arms))
        t += len(arr_arms)
        if t >= T:
            return regret
        
    return regret

def simulate_UCB_second(params, time_range, alpha, delta):
    mean_regret = []
    bandit = GaussianBandit(params)
    pbar = tqdm(time_range)
    regret = dict()
    for t in pbar:
        pbar.set_description("Processing T = %s" % t)
        regret[t] = []
        for i in range(1, 1000 + 1):
            regret[t].append(UCB_second(bandit, T=t, alpha=alpha, delta=delta))

        mean_regret.append(np.mean(regret[t]))
        
    return np.array(mean_regret), regret
```

## MOSS algorithm

```python
def MOSS(bandit, T):
    if T < bandit.K():
        raise Exception("T must be greater than number of arms K.")
        
    n = bandit.K()
    mean_rewards = bandit.pull(np.arange(n))
    n_terms = np.ones(n)
    regret = np.sum(bandit.regret(np.arange(n)))
    arms = np.arange(n)
    for t in range(n, T):
        eq = mean_rewards + np.sqrt(4 / n_terms * np.maximum(1, T / (n * n_terms)))
        mask = eq == np.max(eq)
        arm = np.random.choice(arms[mask])
        mean_rewards[arm] = (mean_rewards[arm] * n_terms[arm] + bandit.pull(arm)) / (n_terms[arm] + 1)
        n_terms[arm] += 1   
        regret += bandit.regret(arm)
        
    return regret

def simulate_MOSS(params, time_range):
    mean_regret = []
    bandit = GaussianBandit(params)
    pbar = tqdm(time_range)
    regret = dict()
    for t in pbar:
        pbar.set_description("Processing T = %s" % t)
        regret[t] = []
        for i in range(1, 1000 + 1):
            regret[t].append(MOSS(bandit, T=t))

        mean_regret.append(np.mean(regret[t]))
        
    return np.array(mean_regret), regret
```

```python
time_range = range(100, 600, 100)
params1 = [1, 3]
params2 = [0.1, 0.5, 1, 3, 5, 7, 9, 12, 15, 20]
mean_regret1, regret1 = simulate_MOSS(params1, time_range)
mean_regret2, regret2 = simulate_MOSS(params2, time_range)
```

## Polynomially Weighted Average Forecaster

Consider polynomial potential
$$
\Phi_p(\vec{u}) = \left( \sum \limits_{i=1}^N (u_i)_+^p \right)^{2/p}, ~ p \geqslant 2,
$$
where $\psi(x) = x^{2/p}, \varphi(x) = x_+^p.$ Moreover
$$
\psi^{\prime}(x) = \dfrac{2}{p} x^{(2 - p) / p}, \\
\varphi^{\prime \prime}(x) = p(p-1)x_+^{p-2}, \\
\Phi(\vec{0}) = 0.
$$
We are going to use the next inequality 
$$
\Phi_p(\vec{R}_n) \leqslant \Phi_p(\vec{0}) + \dfrac{1}{2} \sum \limits_{t=1}^n C(\vec{r}_t)
$$
where
$$
C(\vec{r}_t) = \sup \limits_{\vec{u} \in \mathbb{R}^N} \psi^{\prime} \left( \sum \limits_{i=1}^N \varphi(u_i) \right) \sum \limits_{i=1}^N \varphi^{\prime \prime}(u_i) r_{i,t}^2.
$$
Using Holder's inequality 
$$
\sum \limits_{i=1}^N \varphi^{\prime \prime}(u_i) r_{i, t}^2 = p(p-1) \sum \limits_{i=1}^N (u_i)_+^{p-2} r_{i, t}^2 \leqslant  p(p-1) \left( \sum \limits_{i=1}^N \left[ (u_i)_+^{p-2} \right]^{p/(p-2)} \right)^{(p-2)/p} \left( \sum \limits_{i=1}^N |r_{i,t}|^p \right)^{2/p}.
$$
Now we have
$$
\psi^{\prime} \left( \sum \limits_{i=1}^N \varphi(u_i) \right) \sum \limits_{i=1}^N \varphi^{\prime \prime}(u_i) r_{i,t}^2 \leqslant 2(p-1) \left( \sum \limits_{i=1}^N |r_{i,t}|^p \right)^{2/p}, \\
C(\vec{r}_t) \leqslant 2(p-1) \Vert \vec{r}_t \Vert_p^2
$$
Finally, we get
$$
\Phi_p(\vec{R}_n) = \left( \sum \limits_{i=1}^N (R_{i,n})_+^p \right)^{2/p} \leqslant (p-1) \sum \limits_{t=1}^n \Vert \vec{r}_t \Vert_p^2.
$$
If we suppose that a loss funciton is bounded such that $0 \leqslant \mathcal{L}(x, y) \leqslant 1$ then
$$
\Phi_p(\vec{R}_n) \leqslant n(p-1)N^{2/p}.
$$
For $\forall i = 1, \ldots, N$ 
$$
R_{i, n} \leqslant \left( \sum \limits_{i=1}^N (R_{i,n})_+^p \right)^{1/p} \leqslant \sqrt{n(p-1)N^{2/p}},
$$
or equvalent
$$
\max \limits_{i = 1, \ldots, N} R_{i,n} = \hat{L}_n - \min \limits_{i = 1, \ldots, N} L_{i, n} \leqslant \sqrt{n(p-1)N^{2/p}}.
$$

## Exponentially Weighted Average Forecaster

Consider exponential potential
$$
\Phi_{\mu}(\vec{u}) = \dfrac{1}{\mu} \ln \left( \sum \limits_{i=1}^N e^{\mu u_i} \right).
$$
In this case, the weigths assigned to the experts are 
$$
w_{i, t-1} = \dfrac{e^{\mu R_{i, t-1}}}{\sum \limits_{j=1}^N e^{\mu R_{j, t-1}}} = \dfrac{e^{-\mu L_{i, t-1}}}{\sum \limits_{j=1}^N e^{-\mu L_{j, t-1}}}.
$$
Also we know
$$
\hat{p}_t = \dfrac{\sum \limits_{i=1}^N w_{i, t-1} f_{i,t}}{\sum \limits_{j=1}^N w_{j, t-1}} = \dfrac{\sum \limits_{i=1}^N e^{-\mu L_{i, t-1}} f_{i,t}}{\sum \limits_{j=1}^N e^{-\mu L_{j,t-1}}} = \sum \limits_{i=1}^N w_{i, t-1} f_{i,t}.
$$
Denote $S_t = \sum \limits_{i=1}^N e^{-\mu L_{i,t}}, S_0 = N.$ For $\forall t = 1, \ldots, n$ we have
$$
\dfrac{S_t}{S_{t-1}} = \dfrac{\sum \limits_{i=1}^N e^{-\mu L_{i,t}}}{\sum \limits_{i=1}^N e^{-\mu L_{i,t-1}}} = \{ L_{i, t} = L_{i, t-1} + \mathcal{L}(f_{i,t}, y_t) \} = \dfrac{\sum \limits_{i=1}^N e^{-\mu L_{i,t-1}} e^{-\mu \mathcal{L}(f_{i,t}, y_t)}}{\sum \limits_{i=1}^N e^{-\mu L_{i,t-1}}} = \sum \limits_{i=1}^N w_{i,t-1} e^{-\mu \mathcal{L}(f_{i,t}, y_t)}.
$$
Now we are going to use __Hoeffding inequality.__ Let $X$ be a random variable with $a \leqslant X \leqslant b.$ Then for any $s \in \mathbb{R}$
$$
\mathbb{E}[e^{s(X - \mathbb{E}[X])}] \leqslant \exp \left( \dfrac{s^2}{8}(b-a)^2 \right).
$$
After applying Hoeffding inequality and using convexity of loss function $\mathcal{L}$ in the first argument, we get 
$$
\ln \dfrac{S_t}{S_{t-1}} = \ln \left(\sum \limits_{i=1}^N w_{i,t-1} e^{-\mu \mathcal{L}(f_{i,t}, y_t)} \right) \leqslant -\mu \sum \limits_{i=1}^N w_{i,t-1} \mathcal{L}(f_{i,t}, y_t) + \dfrac{\mu^2}{8} \leqslant - \mu \mathcal{L} \left(\sum \limits_{i=1}^N w_{i,t-1} f_{i,t} , y_t \right) + \dfrac{\mu^2}{8} = - \mu \mathcal{L} \left(\hat{p}_t , y_t \right) + \dfrac{\mu^2}{8}.
$$
Summing over $t = 1, \ldots, n,$ from one hand we have
$$
\ln \dfrac{S_n}{S_{0}} \leqslant - \mu \hat{L}_n + \dfrac{\mu^2 n}{8}.
$$
From the other hand
$$
\ln \dfrac{S_n}{S_{0}} = \ln \left( \sum \limits_{i=1}^N e^{-\mu L_{i,n}} \right) - \ln(N) \geqslant \ln \left( \max \limits_{i = 1, \ldots, N} e^{-\mu L_{i,n}} \right) - \ln(N) \geqslant \{ Jensen \} \geqslant \max \limits_{i = 1, \ldots, N} ( -\mu L_{i,n}) - \ln(N) = - \mu \min \limits_{i = 1, \ldots, N} L_{i,n} - \ln(N).
$$
Finally, we have 
$$
\max \limits_{i = 1, \ldots, N} R_{i, n} = \hat{L}_n - \min \limits_{i = 1, \ldots, N} L_{i, n} \leqslant \dfrac{\ln N}{\mu} + \dfrac{\mu n}{8}.
$$
